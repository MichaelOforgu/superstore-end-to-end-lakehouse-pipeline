```sql
SELECT
  o.order_id,
  o.branch_id,
  CAST(o.date AS DATE) AS order_date,
  o.user_id,
  o.name_surname AS customer_name,
  o.total_basket,
  od.order_detail_id,
  od.amount,
  od.unit_price,
  od.total_price,
  od.item_id,
  c.item_name
FROM dev.bronze.orders o 
INNER JOIN dev.bronze.order_details od
  ON o.order_id = od.order_id
LEFT JOIN dev.bronze.categories c   
  ON od.item_id = c.item_id
WHERE o.order_id = 8510832
```
